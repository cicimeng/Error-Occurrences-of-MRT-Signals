## Modelling

```python
df=df_new
df
```

```python
result = adfuller(df['every_year_each_day_area_cnt'].values)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
df_diff = np.diff(df['every_year_each_day_area_cnt'])
df['every_year_each_day_area_cnt_diff_1'] = np.append([0], df_diff)
df
```

```python
f, ax = plt.subplots(nrows=2, ncols=1, figsize=(16, 8))

plot_acf(df['every_year_each_day_area_cnt_diff_1'], lags=30, ax=ax[0])
plot_pacf(df['every_year_each_day_area_cnt_diff_1'], lags=30, ax=ax[1])

plt.show()
```

```python
train_size = int(0.85 * len(df))
test_size = len(df) - train_size
train_size, test_size
```

```python
univariate_df = df[['Date', 'every_year_each_day_area_cnt']].copy()
univariate_df.columns = ['ds', 'y']
univariate_df
```

```python
train = univariate_df.iloc[:train_size, :]
train
```

```python
x_train, y_train = pd.DataFrame(univariate_df.iloc[:train_size, 0]), pd.DataFrame(univariate_df.iloc[:train_size, 1])
x_valid, y_valid = pd.DataFrame(univariate_df.iloc[train_size:, 0]), pd.DataFrame(univariate_df.iloc[train_size:, 1])
```

```python
print(len(train), len(x_valid))
```

## ARIMA Model

```python
# Fit model
model = ARIMA(y_train, order=(1,1,1))
model_fit=model.fit()

# Prediction with ARIMA
y_pred, se, conf = model_fit.forecast(709,alpha=0.05)

# Calcuate metrics
score_mae = mean_absolute_error(y_valid, y_pred)
score_rmse = math.sqrt(mean_squared_error(y_valid, y_pred))

print('RMSE: {}'.format(score_rmse))
```
